# **Actividad 1: Suma de Vectores**


- Los estudiantes calcularán la suma de vectores que representan el número de entregas realizadas por diferentes conductores en un día específico.
- Actividad: Sumar los vectores para obtener el total de entregas.
- Datos de Entregas:

- Conductores: A, B, C
- Entregas por día (número de paquetes entregados en un día):
- Conductor A: [15, 18, 22, 20, 17]
- Conductor B: [10, 12, 15, 20, 18]
- Conductor C: [25, 30, 28, 27, 29]

```python
import torch as torch

Conductor_A = torch.tensor([15, 18, 22, 20, 17])
Conductor_B = torch.tensor([10, 12, 15, 20, 18])
Conductor_C = torch.tensor([25, 30, 28, 27, 29])

SumaVectores = torch.add((torch.add(Conductor_A,Conductor_B)),Conductor_C)
print("Suma de vectores: ",SumaVectores)
```

Se analizarán las relaciones entre la distancia recorrida por los conductores y el tiempo de entrega.

# **Actividad 6: Distribución de Probabilidad Normal**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 36  # Media
sigma = 5  # Desviación estándar

# Intervalo de tiempo de entrega
# x = np.linspace(30,40,100)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)

# Probabilidad de que el tiempo de entrega esté entre 21 y 51 minutos
probabilidad = stats.norm.cdf(51, mu, sigma) - stats.norm.cdf(21, mu, sigma)
print(f"Probabilidad de que el tiempo de entrega esté entre 21 y 51 minutos: {probabilidad:.4f}")

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal', color='blue')
plt.fill_between(x, y, where=(x >= 21) & (x <= 51), color='skyblue', alpha=0.5) # Mostrar espacio entre numeros de probabilidad calculados
plt.axvline(mu, color='red', linestyle='dashed', label='Media (36 min)')
plt.axvline(21, color='red', linestyle='dashed', label='Límite Inferior (21 min)')
plt.axvline(51, color='green', linestyle='dashed', label='Límite Superior (51 min)')
plt.title('Distribución Normal de Tiempos de Entrega')
plt.xlabel('Tiempo de Entrega (minutos)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid()
plt.show()
```

# **Actividad 7: Distribución de Probabilidad Binomial**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 20  # Número de ensayos
p = 0.7  # Probabilidad de éxito

# Valores de k para calcular la distribución binomial
k_values = np.arange(0, n + 1)
probabilities = binom.pmf(k_values, n, p)

# Cálculo de las probabilidades para k = 15, 10, 5
prob_15 = binom.pmf(15, n, p)
prob_10 = binom.pmf(10, n, p)
prob_5 = binom.pmf(5, n, p)

print(f"P(X=15) = {prob_15:.4f}")
print(f"P(X=10) = {prob_10:.4f}")
print(f"P(X=5) = {prob_5:.4f}")

# Gráfico de la distribución binomial
plt.bar(k_values, probabilities, color='blue', alpha=0.7, label='Distribución Binomial')
plt.scatter([15, 10, 5], [prob_15, prob_10, prob_5], color='red', zorder=3, label='Valores Destacados')
plt.xlabel('Número de Entregas a Tiempo')
plt.ylabel('Probabilidad')
plt.title('Distribución Binomial (n=20, p=0.7)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axhline(y=prob_15, color='r', linestyle='--', label=f"15 entregas")
plt.axhline(y=prob_10, color='b', linestyle='--', label=f"10 entregas")
plt.axhline(y=prob_5, color='g', linestyle='--', label=f"5 entregas")
plt.show()

```
